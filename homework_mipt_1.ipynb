```python
def QuickSort(A):
    if len(A) <= 1:
        return A
    else:
        q = random.choice(A)
        L = [elem for elem in A if elem < q]
        M = [q] * A.count(q)
        R = [elem for elem in A if elem > q] 
        return QuickSort(L) + M + QuickSort(R)
```

```python
def fib_d(n): #динамическая
    a = 0
    b = 1
    for i in range(n):
        a, b = b, a + b
    return a
```

```python
def fib_r(n): #рекурсия
    if n in (1, 2):
        return 1
    return fib_r(n - 1) + fib_r(n - 2)
```

Что такое git и чем он отличается от GitHub?

Git — распределённая система контроля версий, которая даёт возможность разработчикам отслеживать изменения в файлах и работать над одним проектом совместно с коллегами.

GitHub — сервис онлайн-хостинга репозиториев, обладающий всеми функциями распределённого контроля версий и функциональностью управления исходным кодом — всё, что поддерживает Git.


Зачем использовать git и подобные ему системы в больших проектах?

система контроля версий сохраняет все версии проекта, поэтому если в большом проекте, где код обычно делится на части и пишется разными людьми, кто-то сделал ошибку, можно откатить назад и найти ошибку

Какие основные термины используются при работе с git?

Fork — копия репозитория. 
Clone — скачивание репозитория с удалённого сервера на локальный компьютер в определённый каталог для дальнейшей работы с этим каталогом как с репозиторием.
Branch — это параллельная версия репозитория. Она включена в этот репозиторий, но не влияет на главную версию, тем самым позволяя свободно работать в параллельной. 
Master — главная или основная ветка репозитория.
Commit — фиксация изменений или запись изменений в репозиторий.
Pull — получение последних изменений с удалённого сервера репозитория.
Push — отправка всех неотправленных коммитов на удалённый сервер репозитория.
Pull Request — запрос на слияние форка репозитория с основным репозиторием. 
Merge — слияние изменений из какой-либо ветки репозитория с любой веткой этого же репозитория.

В каких случаях имеет смысл создавать различные ветки проекта?

Если мы поняли, что в программе есть баг, но не знаем, где конкретно, на параллельной ветке(branch) мы можем проверить каждый commit на баги и найти ошибку, чтобы не затронуть master, а потом сделать слияние ветки с исправленными багами и основной. 


В чем отличия merge и rebase? В чем преимущества и недостатки последнего из них?

Они переносят commitы из одной ветки в другую, но по-разному. С помощью merge можно определить кто и в какое время сохранил какие-то изменеyия. Rebase также позволяет сохранять изменения, но мы не можем определить кто и когда внес изменения

В чем отличия централизованных СКВ от децентрализованных?

Централизованная СКВ: вся история изменений лежит на сервере, к которому все имеют доступ, главный файл все время обновляется.

Децентрализованная СКВ: у каждого есть своя локальная копия файла, если сервер отключат, то можно продолжать работать со своей копией файла.
